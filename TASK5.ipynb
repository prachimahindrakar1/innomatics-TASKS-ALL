```python
# assignment 5


# problem 1
def count_likes(data):

    total = 0

    index = 0

    while index < len(data):

        total = total + data[index]

        index = index + 1

    if total >= 1000:

        return total, "Viral Post"

    return total, "Normal Engagement"



# problem 2
def stock_check(value):

    result = ""

    if value < 10:

        result = "Low Stock Alert"

    else:

        result = "Stock Sufficient"

    return result



# problem 3
def rainfall_check(values, limit):

    total = 0

    i = 0

    while i < len(values):

        total = total + values[i]

        i = i + 1

    avg = total / len(values)

    if avg >= limit:

        msg = "Adequate Rainfall"

    else:

        msg = "Inadequate Rainfall"

    return avg, msg



# problem 4
def duplicate_check(arr):

    temp = set()

    flag = "No"

    for item in arr:

        if item in temp:

            flag = "Yes"

        temp.add(item)

    return flag



# problem 5
def age_check(num):

    answer = "Not Eligible"

    if num >= 18:

        answer = "Eligible"

    return answer



# problem 6
def price_check(arr):

    output = []

    for i in arr:

        if i > 2000:

            output = output + [i]

    return output



# problem 7
def error_check(num):

    if num == 0:

        return "Healthy"

    if num <= 5:

        return "Minor Issues"

    return "Critical Issues"



# problem 8
def transaction_check(val):

    limit = 50000

    status = "Rejected"

    if val <= limit:

        status = "Approved"

    return status



# problem 9
def attendance_check(arr):

    present = 0

    for i in range(0, len(arr)):

        if arr[i] == "P":

            present = present + 1

    percent = present * 100 / len(arr)

    if percent >= 75:

        return percent, "Eligible"

    return percent, "Not Eligible"



# problem 10
def bill_check(units):

    bill = 0

    if units > 200:

        bill = 300 + 500 + (units - 200) * 7

    elif units > 100:

        bill = 300 + (units - 100) * 5

    else:

        bill = units * 3

    if bill > 1500:

        status = "High Usage"

    elif bill >= 500:

        status = "Moderate Usage"

    else:

        status = "Low Usage"

    return bill, status
```

```python
bill, status = bill_check(260)

print("Bill:", bill)
print("Status:", status)
```
