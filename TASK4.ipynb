```python
# Internship Task 4 – Functions and Loops


# ==================================================
# Problem 1: Mobile Recharge Validator
# ==================================================

def check_recharge_validity(user_amount):
    # list of allowed recharge plans
    available_plans = [199, 299, 399, 599]

    # condition 1: minimum recharge
    if user_amount < 50:
        return False

    # condition 2: must match one of allowed plans
    if user_amount in available_plans:
        return True
    else:
        return False


def recharge_process():
    print("\n--- Mobile Recharge System ---")
    print("Allowed plans: 199, 299, 399, 599")

    while True:
        amt = int(input("Enter recharge amount: ₹"))

        result = check_recharge_validity(amt)

        if result:
            print("Recharge successful and valid.")
            break
        else:
            print("Invalid recharge amount.")
            choice = input("Try again? (yes/no): ")
            if choice.lower() != "yes":
                print("Recharge cancelled.")
                break


# ==================================================
# Problem 2: Inventory Monitoring System
# ==================================================

def stock_checker(product_data):
    print("\n--- Inventory Status ---")

    for item in product_data:
        quantity = product_data[item]

        if quantity < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


# ==================================================
# Problem 3: Student Result Evaluator
# ==================================================

def evaluate_student(mark_list):

    total_marks = 0

    # calculating total using loop
    for value in mark_list:
        total_marks += value

    average = total_marks / len(mark_list)

    # checking pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# ==================================================
# Problem 4: Cab Fare Calculator
# ==================================================

def fare_calculation(distance, peak_status):

    base = 50
    rate_per_km = 12

    total_cost = base + (distance * rate_per_km)

    if peak_status == True:
        total_cost = total_cost + (total_cost * 0.25)

    return total_cost


def cab_system():
    print("\n--- Cab Fare System ---")

    while True:
        km = float(input("Enter distance travelled (km): "))
        peak = input("Peak hour? (yes/no): ")

        if peak.lower() == "yes":
            peak_flag = True
        else:
            peak_flag = False

        fare = fare_calculation(km, peak_flag)

        print("Total Fare: ₹", fare)

        again = input("Calculate again? (yes/no): ")
        if again.lower() != "yes":
            break


# ==================================================
# Problem 5: Attendance Eligibility System
# ==================================================

def check_attendance(record):

    present_days = 0
    total_days = len(record)

    for status in record:
        if status == "P":
            present_days += 1

    percent = (present_days / total_days) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# ==================================================
# Problem 6: Password Security Checker
# ==================================================

def password_strength(password):

    special_symbols = "@#$"

    digit_found = False
    special_found = False

    # check length
    if len(password) < 8:
        return "Weak Password"

    # checking each character
    for ch in password:

        if ch.isdigit():
            digit_found = True

        if ch in special_symbols:
            special_found = True

    if digit_found and special_found:
        return "Strong Password"
    else:
        return "Weak Password"


# ==================================================
# Example Test Runs (optional)
# ==================================================

# recharge_process()

# example inventory
inventory_data = {
    "Milk": 12,
    "Rice": 30,
    "Soap": 8,
    "Sugar": 20
}

stock_checker(inventory_data)

# example student marks
marks = [60, 55, 70, 40]
print("\nStudent Result:", evaluate_student(marks))

# cab_system()

# attendance example
attendance = ["P", "P", "A", "P", "P"]
print("\nAttendance Status:", check_attendance(attendance))

# password example
print("\nPassword Check:", password_strength("Prachi@123"))
```
