```python
# Dictionary of employees and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
max_score = max(employees.values())

# Find employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Print result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", max_score, ")")
```

```python
# Input sentence
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
query = query.translate(str.maketrans('', '', string.punctuation))

# Split words
words = query.split()

# Count frequency
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display keywords searched more than once
result = {}

for word, count in freq.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through readings with index
for i, value in enumerate(sensor_readings):

    # Check if reading is even
    if value % 2 == 0:
        valid_readings.append((i, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count = {}

# Extract domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percent = (count / total) * 100
    print(domain + ":", int(percent), "%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)

# Detect spikes
for i, value in enumerate(sales):

    if value > avg_sales * 1.3:
        print("Day", i+1, ":", value)
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# Count occurrences
for user in user_ids:
    count[user] = count.get(user, 0) + 1

# Print duplicates only
for user, c in count.items():
    if c > 1:
        print(user, "→", c, "times")
```
