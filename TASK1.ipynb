```python
#1.User Login Check

username = "admin"
password = "1234"
a=input()
b=input()
if username==a and password==b:
  print('Login Successful')
else:
  print('Invalid Credentials')
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
count_pass=count_fail=0
for i in marks:
  if i >=50:
    count_pass+=1
  else:
    count_fail+=1
print('no. of pass student : ',count_pass)
print('no. of fail students : ',count_fail)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
li=[]
for i in names:
  li.append(i.strip().lower())
print(li)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  print(i,':',len(i))
  if len(i)>10:
    print('this message is longer than 10 letters')
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
tot=0
for i in logs:
  if i=='ERROR':
    tot+=1
print(tot)
```
