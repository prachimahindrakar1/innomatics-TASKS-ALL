```python
# 2. Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_name = None
highest_salary = -1

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

print(f"Highest Salary: {highest_name} - {highest_salary}")
```

```python
# 3. Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for x in numbers:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# 4. Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# 5. Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

present_count = 0
for mark in attendance:
    if mark == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100
print("Attendance Percentage:", percentage)
```
