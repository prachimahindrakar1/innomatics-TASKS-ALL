```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full / Over Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):

    base_time = distance * 5   # 5 minutes per km

    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5

    if weather == "Rainy":
        base_time += 5
    elif weather == "Storm":
        base_time += 15

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
def seat_occupancy(total_seats, booked_seats):

    booked = len(booked_seats)

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1]*150

seat_occupancy(total_seats, booked_seats)
```

```python
def server_load(cpu_readings):

    total = 0

    for i in cpu_readings:
        total += i

    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu = [45, 60, 70, 85, 90]

server_load(cpu)
```

```python
def resource_monitor(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


# Sample Input
resources = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

resource_monitor(resources)
```

```python
def event_registration(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
